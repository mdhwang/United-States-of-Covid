```python
def create_geojson_features_modded(df):
    print('> Creating GeoJSON features...')
    features = []
    for _, row in df.iterrows():
        for state in geo_state_features['features']:
            if row.state == state['id']:
                feature = state
                feature['properties']['time'] = row.date.date().__str__()
                feature['properties']['iconstyle'] = {'fillcolor':row['color'],'fillOpacity': 0.8}
        features.append(feature)
    return features
```
